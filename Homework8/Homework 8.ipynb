__Exercise 2.__ Modify the query above to return the gene name in yeast, the gene names in the other three species, the cai value and the sequence for all yeast genes with a cai value larger that 0.8 and that have orthologs in the other three species.

```python
pd.read_sql("SELECT orthologs.*, cai, seq FROM cai, orthologs, sequences WHERE gene_name = cerevisiae and cerevisiae = title and CAI > .8", con)
```

__Exercise 5:__ Download the yeast genome gff file from this link: http://downloads.yeastgenome.org/curation/chromosomal_feature/saccharomyces_cerevisiae.gff

Read about the gff format and write a script that will create a database table with all the yeast genes. Your table should have the following columns: `Accession number, Chromosome number, Start Position, End Position, Strand, Feature, Description`.

All this information should be available form the gff file. Only include in your database the yeast protein coding genes, the tRNA coding genes and the full chromosomes. Your final table should have one row per gene/chromosome for a total of 6916 rows.

Note that the description of several genes, and even some gene names contain special HTML characters, like %20. To convert these to text you should use the `unquote()` function of the `urllib` package. Do that before trying to use your regular expressions.

```python
#start from new file and make gff
filein = open('saccharomyces_cerevisiae.gff')
g = []

import urllib
import re

for line in filein:
    line = urllib.unquote(line)
    if line[0:3] == '###':
        break
    elif line[0] == '#':
        continue
    else:
        temp = line.split() #split returs a string with 9 elements
        if temp[2] in ['gene', 'tRNA_gene', 'chromosome']:
            m = re.search('ID=([-\w\(\)]+);',temp[8])
            acc = m.group(1)
            
            
            g.append([acc, temp[0],temp[3], temp[4], temp[6], temp[2], temp[8]])
pd.DataFrame(g,columns=['Acc','Chrom','Start','End','Strand','Feature','Description'])
```
